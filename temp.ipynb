```python
import pandas as pd
data = pd.read_csv('/home/ubuntu/crawl/crawler-recommend-sys/data/sample_0_100_test.csv')
print(data.iloc[1])
```

```python
df = data
url_counts = df['website_url'].value_counts()
urls_with_enough_samples = url_counts[url_counts >= 2].index
print(f"\nURLs có >= 2 samples: {len(urls_with_enough_samples)}")
print(f"URLs có < 2 samples: {(url_counts < 2).sum()}")

# Lọc lại dataframe
df_filtered = df[df['website_url'].isin(urls_with_enough_samples)].reset_index(drop=True)

# Bước 2: Split sao cho mỗi URL có samples ở cả 2 tập
train_indices = []
test_indices = []

for url in df_filtered['website_url'].unique():
    url_data = df_filtered[df_filtered['website_url'] == url].index.tolist()
    
    # Split samples của URL này
    split_point = max(1, len(url_data) // 2)  # ít nhất 1 sample cho mỗi tập
    
    train_indices.extend(url_data[:split_point])
    test_indices.extend(url_data[split_point:])

df_train = df_filtered.loc[train_indices].reset_index(drop=True)
df_test = df_filtered.loc[test_indices].reset_index(drop=True)

print(f"\n=== Cách 2: Kết quả ===")
print(f"Train size: {len(df_train)}")
print(f"Test size: {len(df_test)}")
print(f"Website URLs trong train: {df_train['website_url'].nunique()}")
print(f"Website URLs trong test: {df_test['website_url'].nunique()}")

# Xác nhận mỗi URL có ở cả 2 tập
urls_train = set(df_train['website_url'])
urls_test = set(df_test['website_url'])
print(f"URLs có ở cả train và test: {len(urls_train & urls_test)}")
print(f"URL chỉ ở train: {urls_train - urls_test}")
print(f"URL chỉ ở test: {urls_test - urls_train}")
```
